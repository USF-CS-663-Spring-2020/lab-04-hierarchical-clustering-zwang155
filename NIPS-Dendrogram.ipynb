```python
import pandas as pd

df = pd.read_csv('NIPS_1987-2015.csv')
df.head()
```

```python
df = df.iloc[:, 1:]
df.head()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(12,10))
dendrogram(Z)
plt.show()
```
